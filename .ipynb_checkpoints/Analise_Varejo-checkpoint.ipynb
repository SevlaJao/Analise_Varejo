# 4. Análise univariada

```python
categoria_pedidos = df.groupby('Category')['Quantity'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (8, 6))
plt.bar(categoria_pedidos.index, categoria_pedidos)
plt.title('Total de pedidos por categoria')
plt.xlabel('Categoria')
plt.ylabel('Pedidos')
plt.show()
```

```python
segmento_pedidos = df.groupby('Segment')['Quantity'].sum()
```

```python
plt.figure(figsize = (8, 6))
plt.bar(segmento_pedidos.index, segmento_pedidos)
plt.title('Total pedidos por segmento')
plt.xlabel('Segmento')
plt.ylabel('Pedidos')
plt.show()
```

```python
subcat_pedidos = df.groupby('Sub-Category')['Quantity'].sum().sort_values()
```

```python
plt.figure(figsize = (8, 6))
plt.barh(subcat_pedidos.index, subcat_pedidos)
plt.title('Total de pedidos por sub-categoria')
plt.xlabel('Pedidos')
plt.ylabel('Sub-categoria')
plt.show()
```

#### Com estes gráficos podemos ver qual categoria, sub-categoria e segmento têm mais pedidos

# 5. Análise bivariada

```python
vendas_ano2 = df.groupby('Ano')['Sales'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize = (8, 6))
plt.bar(vendas_ano2.index, vendas_ano2)
plt.title('Total de vendas por ano')
plt.xlabel('Ano')
plt.ylabel('Total de vendas')
plt.show()
```

```python
vendas_estado = df.groupby('State')['Sales'].sum().sort_values(ascending = True)
```

```python
plt.figure(figsize = (10, 8))
plt.barh(vendas_estado.index, vendas_estado)
plt.title('Total de vendas por estado')
plt.xlabel('Vendas')
plt.ylabel('Esstado')
plt.show()
```

```python
cidade_vendas = df.groupby('City')['Sales'].sum().sort_values(ascending = False)
```

```python
cidade_vendas = cidade_vendas[:10].sort_values(ascending = True)
```

```python
plt.figure(figsize = (8, 6))
plt.barh(cidade_vendas.index, cidade_vendas)
plt.title('As 10 cidades com mais vendas')
plt.xlabel('Vendas')
plt.ylabel('Cidades')
plt.show()
```

```python
segmento_vendas = df.groupby('Segment')['Sales'].sum()
```

```python
plt.figure(figsize = (8, 6))
plt.bar(segmento_vendas.index, segmento_vendas)
plt.title('Total de vendas por segmento')
plt.xlabel('Segmento')
plt.ylabel('Vendas em escala de milhão')
plt.show()
```

#### Com essas informações podemos ver o total de vendas por ano, estado,cidades e segmento

# 6. Análise multivariada

```python
cidade_categoria = df[df['Category'] == 'Office Supplies']
```

```python
cidade_categoria_vendas = cidade_categoria.groupby('City')['Sales'].sum().sort_values(ascending = False).reset_index()
```

```python
cidade_categoria_vendas = cidade_categoria_vendas[:10]
```

```python
plt.figure(figsize = (8, 6))
plt.barh(cidade_categoria_vendas['City'], cidade_categoria_vendas['Sales'])
plt.title('Total de vendas da categoria Office Supplies por cidade ')
plt.xlabel('Vendas')
plt.ylabel('Cidades')
plt.show()
```

```python
vendas_segmento_ano = df.groupby(['Ano', 'Segment'])['Sales'].sum().unstack()
```

```python
largura_barra = 0.2
```

```python
posicoes = np.arange(len(vendas_segmento_ano.index))
```

```python
plt.figure(figsize = (8, 6))
plt.bar(posicoes - 1.5 * largura_barra, vendas_segmento_ano['Consumer'], width = largura_barra, label = 'Consumer')
plt.bar(posicoes - 0.5 * largura_barra, vendas_segmento_ano['Corporate'], width = largura_barra, label = 'Corporate')
plt.bar(posicoes + 0.5 * largura_barra, vendas_segmento_ano['Home Office'], width = largura_barra, label = 'Home Office')
plt.xticks(posicoes, vendas_segmento_ano.index)
plt.title('Total de vendas por segmento e ano')
plt.legend()
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.show()
```
